# Single proper experiment with default ModelSpecs , can tune this accordingly 

```python
from pydart.experiment import run_experiment

def main() -> None:
    evaluator = run_experiment(
        k=2,
        heavy_light_ratio=(1, 1),
        num_tasks=4,
    )

    print("\n=== Summary ===")
    # print("For the ratio")
    print(f"Naive makespan: {evaluator.naive_makespan:.4f}s")
    print(f"Parallel makespan: {evaluator.parallel_makespan:.4f}s")
    print(f"Speedup: {evaluator.speedup_makespan:.2f}x")
    print(f"Parallel throughput: {evaluator.throughput_makespan:.2f} tasks/s")


if __name__ == "__main__":
    main()
```
